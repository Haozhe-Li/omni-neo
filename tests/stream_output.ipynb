```python
def calculator(a: int, b: int, op: str) -> float:
    """Basic calculator

    Args:
        a (int): first number
        b (int): second number
        op (str): operator (+, -, *, /)

    Returns:
        float: result
    """
    if op == "+":
        return a + b
    elif op == "-":
        return a-b
    elif op == "*":
        return a*b
    else:
        return a/b
```

```python
agent = create_agent(
    model=model,
    tools = [calculator]
)
```
